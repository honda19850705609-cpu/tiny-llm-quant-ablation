```python
%cd /content/drive/MyDrive/AI_learning/tiny-llm-quant-ablation/github
import torch, json
from tllm.model import TinyLLM, ModelConfig
from tokenizers import Tokenizer

device = 'cuda' if torch.cuda.is_available() else 'cpu'
ck = torch.load('/content/drive/MyDrive/AI_learning/tiny-llm-quant-ablation/model/ckpt.pt', map_location=device)
cfg = ModelConfig(**ck['cfg'])
model = TinyLLM(cfg).to(device); model.load_state_dict(ck['model']); model.eval()

tok = Tokenizer.from_file('/content/drive/MyDrive/AI_learning/tiny-llm-quant-ablation/tllm-data/tokenizer.json')

prompt = "Once upon a time, there was a little"
ids = torch.tensor([tok.encode(prompt).ids], device=device)
out = model.generate(ids, max_new_tokens=100, temperature=0.8, top_k=200)
print(tok.decode(out[0].tolist()))
```
